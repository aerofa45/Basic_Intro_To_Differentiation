### Explanation of Gradient Descent in Neural Networks  

Neural networks learn by **minimizing errors** in their predictions. This is done using a method called **gradient descent**, which relies on differentiation to update the network’s **weights** (parameters that control how inputs are transformed to outputs).  


#### Understanding the Formula for Weight Update  
The weight update rule in gradient descent is:

Where:
- w is the **weight** (a numerical value that determines how much influence an input has on the output).  
- η (eta) is the **learning rate** (a small positive number that controls how big of a step we take when updating weights).  
- L is the **loss function** (a measure of how far off the predictions are from actual values).  
- ∂L/∂w is the **gradient** (the derivative of the loss function with respect to the weight, showing how much the loss changes when the weight changes).  

This formula means we adjust the weight in the opposite direction of the gradient to decrease the loss, thereby improving the model’s accuracy.  











 How It Works  
1. Start with an initial guess for the parameters (weights) you want to optimize.  
2. Calculate the gradient (derivative) of the function with respect to the parameters.  
3. Update the parameters by moving them a little bit in the opposite direction of the gradient.  
4. Repeat the process until the gradient becomes small enough (close to zero), meaning you've reached a minimum.  

---

###  Mathematically,  
If we want to minimize a function `f(x)`, we update `x` using:

```
x_new = x_old - α * ∇f(x)
```

Where:  
- α = Learning rate (a small positive value that controls the step size).  
- ∇f(x) = Gradient of the function (partial derivatives of `f(x)` with respect to all parameters).  



For Ex, we want to minimize:

```
f(x) = x^2
```

The derivative is:  

```
d/dx (x^2) = 2x
```

Using gradient descent, the update rule is:  

```
x_new = x_old - α * (2x)
```

---

Visualization  
Imagine you're standing on a hill and you want to reach the lowest point (valley).  
- The gradient tells you which direction is the steepest upward.  
- By moving in the opposite direction, you keep moving downhill toward the lowest point.

In neural networks, gradient descent is used to **adjust the weights and biases** during training so that the network learns to minimize the error (loss function) and make better predictions.  


Gradient descent implementation:

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return x**2 + 2*x

def f_derivative(x):
    return 2 * x

# Gradient Descent function
def gradient_descent(learning_rate, initial_x, iterations):
    x = initial_x
    x_history = [x]  # To store the x values during the process

    for i in range(iterations):
        grad = f_derivative(x)               # Calculate gradient (slope)
        x = x - learning_rate * grad         # Update x using gradient descent
        x_history.append(x)                  # Store the new x value

    return x, x_history

# Parameters
learning_rate = 0.1
initial_x = 10
iterations = 20

# Run Gradient Descent
final_x, x_history = gradient_descent(learning_rate, initial_x, iterations)

# Plotting the process
x_values = np.linspace(-12, 12, 100)
y_values = f(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='$f(x) = x^2 + 2*x$', color='blue')
plt.scatter(x_history, [f(x) for x in x_history], color='red')
plt.plot(x_history, [f(x) for x in x_history], color='red', linestyle='dashed', label='Gradient Descent Path')
plt.title('Gradient Descent Optimization Process')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# Print the final value of x and f(x)
print(f"Final x value: {final_x}")
print(f"Final f(x) value: {f(final_x)}")
````



#  Building a Simple Neural Network from Scratch



---

#### Neural Network?
A neural network consists of layers of neurons, where each neuron performs a weighted sum of inputs followed by an activation function:  

$$  
z = w \cdot x + b  
$$  

where:  
- $w$ is the weight  
- $x$ is the input  
- $b$ is the bias  

The output is transformed using an activation function, such as the **sigmoid function**, which introduces non-linearity:  

$$  
a = \sigma(z) = \frac{1}{1 + e^{-z}}  
$$  

This activation function ensures that the output is between 0 and 1.

---

#### Forward Pass: Computing the Output
For a given input $x$, the neuron computes:  

$$  
z = w \cdot x + b  
$$  

$$  
a = \sigma(z) = \frac{1}{1 + e^{-z}}  
$$  

where $\sigma(z)$ is the sigmoid function.

---

#### Backward Pass: Learning with Differentiation
To update weights and bias, we use **gradient descent**, which requires differentiation. The goal is to minimize the error using the derivative of the loss function.

#### Error Calculation (Loss Function)
We define the **error** as the difference between the predicted output $a$ and the actual label $y$:  

$$  
\text{Error} = a - y  
$$  

#### Computing Gradients (Differentiation)
To update the weights and bias, we need their gradients:

1. **Derivative of the Sigmoid Function**:  
   Since we use the sigmoid activation function, its derivative is:  

   $$  
   \sigma'(z) = \sigma(z)(1 - \sigma(z))  
   $$  

2. **Weight Gradient**:  
   The partial derivative of the error with respect to $w$ is:  

   $$  
   \frac{\partial \text{Error}}{\partial w} = (\text{Error}) \cdot x \cdot \sigma'(z)  
   $$  

3. **Bias Gradient**:  
   The partial derivative of the error with respect to $b$ is:  

   $$  
   \frac{\partial \text{Error}}{\partial b} = (\text{Error}) \cdot \sigma'(z)  
   $$  

These gradients help adjust $w$ and $b$ to minimize the error.

---

## Summary
- The forward pass computes $z$ and applies the sigmoid function to obtain the output.  
- The backward pass uses differentiation to compute gradients and update the weight and bias.  
- The gradient descent algorithm iteratively adjusts $w$ and $b$ to minimize the error.  



````python


import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of the sigmoid function
def sigmoid_derivative(x):
    return x * (1 - x)

# Initialize parameters
weights = np.array([0.5])  # Initial weight
bias = 0.0
learning_rate = 0.1

# Training data
X = np.array([1, 2, 3])  # Inputs
Y = np.array([0, 1, 1])  # Labels

# Training loop
for epoch in range(10):
    for i in range(len(X)):
        # Forward pass
        z = X[i] * weights + bias
        pred = sigmoid(z)

        # Compute error
        error = pred - Y[i]

        # Compute gradients using differentiation
        dW = error * X[i] * sigmoid_derivative(pred)
        dB = error * sigmoid_derivative(pred)

        # Update weights using gradient descent
        weights -= learning_rate * dW
        bias -= learning_rate * dB

    print(f'Epoch {epoch}, Weight: {weights}, Bias: {bias}')
````

### Partial Derivative Explanation

A **partial derivative** measures how a function changes when we tweak **only one variable at a time**, while keeping all other variables **fixed**.

#### Example:
Imagine you're baking cookies. The taste of the cookies depends on **two things**:
- The amount of **sugar** (S)
- The amount of **flour** (F)

We can think of the taste as a function:  
\[
T = f(S, F)
\]
where **T** is the taste, and it changes based on how much sugar (S) and flour (F) we use.

#### What is a Partial Derivative?
If we take the **partial derivative with respect to sugar (S)**, we’re asking:
*"How does the taste change if I add a little more sugar, but keep the flour exactly the same?"*

Mathematically, we write this as:
$$
\frac{\partial T}{\partial S}
$$
which reads as "the partial derivative of taste with respect to sugar."

Similarly, if we take the **partial derivative with respect to flour (F)**, we’re asking:
*"How does the taste change if I add more flour, but don’t touch the sugar?"*

Mathematically, we write:
$$
\frac{\partial T}{\partial F}
$$

#### Simple Math Example

Consider the function:
$$
f(x, y) = 3x^2 + 2y
$$
where **x** and **y** are variables, and **f(x, y)** gives the output.

##### 1. Partial Derivative with respect to x:
To find the partial derivative of \( f(x, y) \) with respect to **x**, we treat **y** as a constant. We only focus on the part of the function that has **x** in it.

The function is:  
$$
f(x, y) = 3x^2 + 2y
$$

Now, take the derivative of each term with respect to **x**:
- The derivative of 3x^2 wiith respect to **x** is  6x  (using the power rule).
- The term 2y does not have an **x** in it, so its derivative with respect to **x** is **0**.

Thus, the partial derivative with respect to **x** is:
$$
\frac{\partial f}{\partial x} = 6x
$$

##### 2. Partial Derivative with respect to y:
Now, let's find the partial derivative of f(x, y) with respect to **y**, treating **x** as a constant.

The function is:
$$
f(x, y) = 3x^2 + 2y
$$

Now, take the derivative of each term with respect to **y**:
- The term 3x^2 does not have a **y** in it, so its derivative with respect to **y** is **0**.
- The derivative of 2y  with respect to **y** is simply 2.

Thus, the partial derivative with respect to **y** is:
$$
\frac{\partial f}{\partial y} = 2
$$

This shows how the function f(x, y) changes when we change **x** or **y**, holding the other variable constant.
